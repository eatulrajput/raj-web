```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
ipl_auction_df = pd.read_csv('IPL.csv')

# Set the plot style
sns.set(style="whitegrid")

# Create a box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='AGE', y='SOLD PRICE', hue='PLAYING ROLE', data=ipl_auction_df)

# Customize the plot
plt.title('Sold Price Distribution by Age and Playing Role')
plt.xlabel('Age')
plt.ylabel('Sold Price')
plt.legend(title='Playing Role', bbox_to_anchor=(1.05, 1), loc=2)

# Show the plot
plt.show()
```
